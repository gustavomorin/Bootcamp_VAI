## 1. Entendimento do Contexto



**Panorama geral**  
A VAI Store é uma rede de varejo alimentar especializada em produtos naturais e saudáveis, com lojas físicas no Sudeste e presença no e-commerce. A demanda por alimentação saudável está em alta, mas a empresa enfrenta desafios para usar os dados de vendas de forma estratégica e preditiva. **Por que é um problema agora?**  
O crescimento do mercado e a necessidade de diferenciar a experiência de compra do cliente, aliado ao grande volume de dados de vendas, tornam imprescindível uma previsão de demanda precisa. Um modelo confiável pode reduzir custos com estoques desnecessários e aumentar a disponibilidade de produtos. **Qual é a questão central?**  
Como podemos, a partir dos dados históricos de vendas e do perfil das lojas, criar um modelo preditivo para prever a demanda de produtos e usar esses insights para melhorar as decisões de compra e vendas da VAI Store?

**Contexto operacional, histórico e organizacional:**

- **Organizacional:** O CEO é analítico e busca transformar a operação com autonomia para a equipe, e quer ver resultados concretos para apoiar a tomada de decisão.

**Quem são as pessoas envolvidas?**  
- **CEO:** Toma decisões com base em dados e espera soluções robustas.  
- **Time de Compras:** Precisa de previsões confiáveis para negociar melhor com fornecedores.  
- **Time de Análise:** Responsável por coletar dados, gerar modelos e apresentar os resultados de forma clara e convincente.

**Quais os interesses envolvidos?**  
- **Melhorar a acurácia das previsões de demanda** para reduzir rupturas e excessos de estoque.  
- **Aumentar a eficiência operacional** com dados preditivos e decisões mais rápidas.  
- **Aprimorar a experiência do cliente** com produtos sempre disponíveis e de acordo com a demanda real.

**Enunciado SMART:**  
"Desenvolver um modelo de previsão de demanda para a VAI Store, utilizando dados históricos de vendas de 2024, que seja capaz de prever a demanda diária de cada produto com acurácia superior a 85% para todo o ano de 2025, subsidiando o time de Compras na definição de tamanhos de lote e apoiando as estratégias de vendas."

```python
# 1️ Remover linhas sem SKU em df_produto
df_produto = df_produto.dropna(subset=['SKU'])

# 2️ Garantir que as colunas SKU sejam do tipo inteiro
df_produto['SKU'] = df_produto['SKU'].astype(int)
df_vendas['SKU'] = df_vendas['SKU'].astype(int)

# 3️ Fazer o merge para criar o dataframe final df
df = df_vendas.merge(df_produto, on='SKU', how='left', suffixes=('', '_prod'))

# 4️ Caso exista alguma coluna SKU_prod duplicada, remover
df = df.drop(columns=['SKU_prod'], errors='ignore')

# 5️ Visualizar as primeiras linhas do dataframe final
df.head()

```

```python
# Garantir que DATA_ATEND está em formato datetime
df['DATA_ATEND'] = pd.to_datetime(df['DATA_ATEND'])

# Mês
df['MES'] = df['DATA_ATEND'].dt.month

# Trimestre
df['TRIMESTRE'] = df['DATA_ATEND'].dt.quarter

# Bimestre (1º bimestre: jan-fev, etc.)
df['BIMESTRE'] = ((df['MES'] - 1) // 2) + 1

# Semestre (1º semestre: jan-jun, 2º semestre: jul-dez)
df['SEMESTRE'] = ((df['MES'] - 1) // 6) + 1

# Estação do ano (Brasil - Hemisfério Sul)
def estacao_do_ano(mes):
    if mes in [12, 1, 2]:
        return 'Verão'
    elif mes in [3, 4, 5]:
        return 'Outono'
    elif mes in [6, 7, 8]:
        return 'Inverno'
    else:
        return 'Primavera'

df['ESTACAO'] = df['MES'].apply(estacao_do_ano)

# Dia da semana (nome do dia)
df['DIA_SEMANA'] = df['DATA_ATEND'].dt.day_name()

# Número do dia da semana (segunda=0, domingo=6)
df['DIA_SEMANA_NUM'] = df['DATA_ATEND'].dt.weekday

# Semana do ano
df['SEMANA_DO_ANO'] = df['DATA_ATEND'].dt.isocalendar().week

# Final de semana (True para sábado e domingo)
df['FIM_DE_SEMANA'] = df['DIA_SEMANA_NUM'] >= 5

# Exibir as primeiras linhas para checar as novas colunas
df.head(100)
```

```python
import matplotlib.pyplot as plt

# Lista de colunas sazonais que você criou
colunas_sazonais = [
    'MES', 'TRIMESTRE', 'BIMESTRE', 'SEMESTRE',
    'ESTACAO', 'DIA_SEMANA', 'SEMANA_DO_ANO', 'FIM_DE_SEMANA'
]

for coluna in colunas_sazonais:
    plt.figure(figsize=(8,4))
    df[coluna].value_counts().sort_index().plot(kind='bar')
    plt.title(f'Distribuição da coluna: {coluna}')
    plt.ylabel('Frequência')
    plt.xlabel(coluna)
    plt.xticks(rotation=45)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ Agrupamento por tipo de filial
df_filial = df.groupby('FILIAL').agg({
    'QTD_VENDA': 'sum',
    'FATUR_VENDA': 'sum'
}).reset_index()

print(df_filial)

# 2️⃣ Gráfico de barras para Quantidade Vendida
plt.figure(figsize=(10,5))
plt.bar(df_filial['FILIAL'], df_filial['QTD_VENDA'], color='skyblue')
plt.xlabel('Tipo de Filial')
plt.ylabel('Quantidade Vendida')
plt.title('Quantidade Vendida por Tipo de Filial')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3️⃣ Gráfico de barras para Faturamento
plt.figure(figsize=(10,5))
plt.bar(df_filial['FILIAL'], df_filial['FATUR_VENDA'], color='lightgreen')
plt.xlabel('Tipo de Filial')
plt.ylabel('Faturamento')
plt.title('Faturamento por Tipo de Filial')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ Agrupamento por categoria
df_categoria = df.groupby('CATEGORIA').agg({
    'QTD_VENDA': 'sum',
    'FATUR_VENDA': 'sum'
}).reset_index()

print(df_categoria)

# 2️⃣ Gráfico de barras para Quantidade Vendida por Categoria
plt.figure(figsize=(10,5))
plt.bar(df_categoria['CATEGORIA'], df_categoria['QTD_VENDA'], color='skyblue')
plt.xlabel('Categoria')
plt.ylabel('Quantidade Vendida')
plt.title('Quantidade Vendida por Categoria')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3️⃣ Gráfico de barras para Faturamento por Categoria
plt.figure(figsize=(10,5))
plt.bar(df_categoria['CATEGORIA'], df_categoria['FATUR_VENDA'], color='lightgreen')
plt.xlabel('Categoria')
plt.ylabel('Faturamento')
plt.title('Faturamento por Categoria')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
