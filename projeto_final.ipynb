## 1. Entendimento do Contexto



**Panorama geral**  
A VAI Store  uma rede de varejo alimentar especializada em produtos naturais e saudveis, com lojas fsicas no Sudeste e presena no e-commerce.
```python
# 1 Remover linhas sem SKU em df_produto
df_produto = df_produto.dropna(subset=['SKU'])

# 2 Garantir que as colunas SKU sejam do tipo inteiro
df_produto['SKU'] = df_produto['SKU'].astype(int)
df_vendas['SKU'] = df_vendas['SKU'].astype(int)

# 3 Fazer o merge para criar o dataframe final df
df = df_vendas.merge(df_produto, on='SKU', how='left', suffixes=('', '_prod'))

# 4 Caso exista alguma coluna SKU_prod duplicada, remover
df = df.drop(columns=['SKU_prod'], errors='ignore')

# 5 Visualizar as primeiras linhas do dataframe final
df.head()

```

```python
# Garantir que DATA_ATEND est em formato datetime
df['DATA_ATEND'] = pd.to_datetime(df['DATA_ATEND'])

# Ms
df['MES'] = df['DATA_ATEND'].dt.month

# Trimestre
df['TRIMESTRE'] = df['DATA_ATEND'].dt.quarter

# Bimestre (1 bimestre: jan-fev, etc.)
df['BIMESTRE'] = ((df['MES'] - 1) // 2) + 1

# Semestre (1 semestre: jan-jun, 2 semestre: jul-dez)
df['SEMESTRE'] = ((df['MES'] - 1) // 6) + 1

# Estao do ano (Brasil - Hemisfrio Sul)
def estacao_do_ano(mes):
    if mes in [12, 1, 2]:
        return 'Vero'
    elif mes in [3, 4, 5]:
        return 'Outono'
    elif mes in [6, 7, 8]:
        return 'Inverno'
    else:
        return 'Primavera'

df['ESTACAO'] = df['MES'].apply(estacao_do_ano)

# Dia da semana (nome do dia)
df['DIA_SEMANA'] = df['DATA_ATEND'].dt.day_name()

# Nmero do dia da semana (segunda=0, domingo=6)
df['DIA_SEMANA_NUM'] = df['DATA_ATEND'].dt.weekday

# Semana do ano
df['SEMANA_DO_ANO'] = df['DATA_ATEND'].dt.isocalendar().week

# Final de semana (True para sbado e domingo)
df['FIM_DE_SEMANA'] = df['DIA_SEMANA_NUM'] >= 5

# Exibir as primeiras linhas para checar as novas colunas
df.head(100)
```

```python
import matplotlib.pyplot as plt

# Lista de colunas sazonais que voc criou
colunas_sazonais = [
    'MES', 'TRIMESTRE', 'BIMESTRE', 'SEMESTRE',
    'ESTACAO', 'DIA_SEMANA', 'SEMANA_DO_ANO', 'FIM_DE_SEMANA'
]

for coluna in colunas_sazonais:
    plt.figure(figsize=(8,4))
    df[coluna].value_counts().sort_index().plot(kind='bar')
    plt.title(f'Distribuio da coluna: {coluna}')
    plt.ylabel('Frequncia')
    plt.xlabel(coluna)
    plt.xticks(rotation=45)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1 Agrupamento por tipo de filial
df_filial = df.groupby('FILIAL').agg({
    'QTD_VENDA': 'sum',
    'FATUR_VENDA': 'sum'
}).reset_index()

print(df_filial)

# 2 Grfico de barras para Quantidade Vendida
plt.figure(figsize=(10,5))
plt.bar(df_filial['FILIAL'], df_filial['QTD_VENDA'], color='skyblue')
plt.xlabel('Tipo de Filial')
plt.ylabel('Quantidade Vendida')
plt.title('Quantidade Vendida por Tipo de Filial')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3 Grfico de barras para Faturamento
plt.figure(figsize=(10,5))
plt.bar(df_filial['FILIAL'], df_filial['FATUR_VENDA'], color='lightgreen')
plt.xlabel('Tipo de Filial')
plt.ylabel('Faturamento')
plt.title('Faturamento por Tipo de Filial')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
