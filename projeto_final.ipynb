```python
# 1️ Remover linhas sem SKU em df_produto
df_produto = df_produto.dropna(subset=['SKU'])

# 2️ Garantir que as colunas SKU sejam do tipo inteiro
df_produto['SKU'] = df_produto['SKU'].astype(int)
df_vendas['SKU'] = df_vendas['SKU'].astype(int)

# 3️ Fazer o merge para criar o dataframe final df
df = df_vendas.merge(df_produto, on='SKU', how='left', suffixes=('', '_prod'))

# 4️ Caso exista alguma coluna SKU_prod duplicada, remover
df = df.drop(columns=['SKU_prod'], errors='ignore')

# 5️ Visualizar as primeiras linhas do dataframe final
df.head()

```

```python
import holidays

# Garantir que DATA_ATEND está em formato datetime
df['DATA_ATEND'] = pd.to_datetime(df['DATA_ATEND'])

# Mês
df['MES'] = df['DATA_ATEND'].dt.month

# Trimestre
df['TRIMESTRE'] = df['DATA_ATEND'].dt.quarter

# Bimestre (1º bimestre: jan-fev, etc.)
df['BIMESTRE'] = ((df['MES'] - 1) // 2) + 1

# Semestre (1º semestre: jan-jun, 2º semestre: jul-dez)
df['SEMESTRE'] = ((df['MES'] - 1) // 6) + 1

# Estação do ano (Brasil - Hemisfério Sul)
def estacao_do_ano(mes):
    if mes in [12, 1, 2]:
        return 'Verão'
    elif mes in [3, 4, 5]:
        return 'Outono'
    elif mes in [6, 7, 8]:
        return 'Inverno'
    else:
        return 'Primavera'

df['ESTACAO'] = df['MES'].apply(estacao_do_ano)

# Dia da semana (nome do dia)
df['DIA_SEMANA'] = df['DATA_ATEND'].dt.day_name()

# Número do dia da semana (segunda=0, domingo=6)
df['DIA_SEMANA_NUM'] = df['DATA_ATEND'].dt.weekday

# Semana do ano
df['SEMANA_DO_ANO'] = df['DATA_ATEND'].dt.isocalendar().week

# Final de semana (True para sábado e domingo)
df['FIM_DE_SEMANA'] = df['DIA_SEMANA_NUM'] >= 5

# Gerar lista de feriados no Brasil para 2024
br_feriados = holidays.Brazil(years=[2024])  

# Adicionar coluna indicando se a data é feriado
df['FERIADO'] = df['DATA_ATEND'].isin(br_feriados)

# Início do mês (primeiros 7 dias)
df['INICIO_MES'] = df['DATA_ATEND'].dt.day <= 7

# Fim do mês (últimos 7 dias)
df['FIM_MES'] = df['DATA_ATEND'].dt.day >= 25

# Para cada SKU, calcula a média móvel de 7 dias do faturamento
df['FATURAMENTO_MM7'] = df.groupby('SKU')['FATUR_VENDA'].transform(lambda x: x.rolling(7, min_periods=1).mean())

df['FATUR_VENDA_LAG1'] = df.groupby('SKU')['FATUR_VENDA'].shift(1)

# Exibir as primeiras linhas para checar as novas colunas
df.head(100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Lista de colunas sazonais
#sazonalidades = ['MES', 'TRIMESTRE', 'BIMESTRE', 'SEMESTRE', 'ESTACAO', 'DIA_SEMANA', 'SEMANA_DO_ANO', 'DATA_ATEND']
sazonalidades = ['SEMANA_DO_ANO']

# Dicionário de ordem personalizada
ordem_personalizada = {
    'ESTACAO': ['Verão', 'Outono', 'Inverno', 'Primavera'],
    'DIA_SEMANA': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
}

for sazonalidade in sazonalidades:
    print(f"\nAnalisando sazonalidade: {sazonalidade}")
    
    df_sazonalidade = df.groupby(['CATEGORIA', sazonalidade])['QTD_VENDA'].sum().reset_index()
    
    # Ajustar ordem personalizada
    if sazonalidade in ordem_personalizada:
        df_sazonalidade[sazonalidade] = pd.Categorical(
            df_sazonalidade[sazonalidade],
            categories=ordem_personalizada[sazonalidade],
            ordered=True
        )
        df_sazonalidade = df_sazonalidade.sort_values(sazonalidade)
    
    for categoria in df_sazonalidade['CATEGORIA'].unique():
        dados_categoria = df_sazonalidade[df_sazonalidade['CATEGORIA'] == categoria].reset_index(drop=True)
        
        x_labels = dados_categoria[sazonalidade]
        y = dados_categoria['QTD_VENDA']
        
        if sazonalidade == 'DATA_ATEND':
            # Converter datas em número de dias desde a primeira data
            x_numeric = (x_labels - x_labels.min()).dt.days
            x_labels_fmt = x_labels.dt.strftime('%Y-%m-%d')
        else:
            if x_labels.dtype.name == 'category' or x_labels.dtype == object:
                x_numeric = np.arange(1, len(x_labels) + 1)
            else:
                x_numeric = x_labels.values
            x_labels_fmt = x_labels
        
        plt.figure(figsize=(10,4))
        plt.plot(x_numeric, y, marker='o', color='steelblue', label='Quantidade Vendida')
        
        # Calcular e plotar linha de tendência polinomial (grau 10)
        grau = 10
        if len(x_numeric) > grau:
            z = np.polyfit(x_numeric, y, grau)
            p = np.poly1d(z)
            plt.plot(x_numeric, p(x_numeric), "r--", label=f'Tendência Polinomial (grau {grau})')
        
        # Destacar os 3 pontos com maior venda
        top3 = dados_categoria.nlargest(3, 'QTD_VENDA')
        x_top3 = [x_numeric[i] for i in top3.index]
        plt.scatter(x_top3, top3['QTD_VENDA'], color='darkorange', s=100, label='Top 3 Sazonalidade')

        # Adicionar rótulos para os Top 3
        for i in range(len(top3)):
            x_val = x_top3[i]
            y_val = top3['QTD_VENDA'].iloc[i]
            label_val = str(top3[sazonalidade].iloc[i])
            plt.text(x_val, y_val, label_val, fontsize=9, ha='center', va='bottom', color='black')

        plt.title(f'{sazonalidade} - Categoria: {categoria}')
        plt.xlabel(sazonalidade)
        plt.ylabel('Quantidade Vendida')
        if sazonalidade == 'DATA_ATEND':
            plt.xticks(x_numeric[::max(len(x_numeric)//10, 1)], x_labels_fmt[::max(len(x_labels_fmt)//10, 1)], rotation=45)
        else:
            if isinstance(x_labels_fmt, pd.Series) or isinstance(x_labels_fmt, pd.Categorical):
                plt.xticks(x_numeric, x_labels_fmt, rotation=45)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

```
